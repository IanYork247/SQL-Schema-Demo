# Introduction to Exploratory Data Analysis: Used Car Pricing

This notebook demonstrates an exploratory data analysis (EDA) of used car pricing data.  
The goal is to identify key factors influencing car prices using Python libraries like Pandas, Matplotlib, Seaborn, and SciPy.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "../python_data/used_car_pricing_data.csv"
df = pd.read_csv(file_path, header=0)

# Display the first 5 rows of the dataset
df.head()
```

## Dataset Overview

The dataset contains information about various attributes of used cars (e.g., engine size, highway MPG, body style, etc.) and their pricing. We'll start by exploring the data structure.

## Exploratory Data Analysis

We'll explore relationships between car attributes and prices using summary statistics, visualizations, and correlation analysis.

```python
# Correlation analysis
df[['engine-size', 'price']].corr()

# Scatterplot: Engine Size vs Price
sns.regplot(x="engine-size", y="price", data=df)
plt.title("Engine Size vs Price")
plt.show()
```

```python
# Boxplot: Body Style vs Price
sns.boxplot(x="body-style", y="price", data=df)
plt.title("Body Style vs Price")
plt.show()
```

## Statistical Analysis

In this section, we'll calculate summary statistics and analyze categorical variables.

```python
# Summary statistics
df.describe()

# Value counts for 'drive-wheels'
df['drive-wheels'].value_counts()
```

## Key Findings

1. **Engine Size**: Strong positive correlation with price.
2. **Drive Wheels**: Rear-wheel drive cars tend to have higher prices.
3. **Highway MPG**: Negative correlation with price, indicating that fuel efficiency affects car value.
4. **Body Style**: Sedans and luxury models generally have higher prices.
